```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('SoccerPlayerStatistics.csv')
```

```python
data.head()
```

```python
data.describe(include='all')
```

```python
data.isna().sum()
```

```python
data.hist(bins=20, figsize=(100, 50))
plt.show()
```

```python
categorical_columns = ['Nationality', 'Club', 'Preffered_Position']
for column in categorical_columns:
    plt.figure(figsize=(15, 8))
    sns.countplot(data=data, x=column, order=data[column].value_counts().index[:20])
    plt.xticks(rotation=90)
    plt.show()
```

```python
columns_to_plot = ['Age']
for column in columns_to_plot:
    plt.figure(figsize=(8, 6))
    sns.histplot(data[column], kde=True)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
data.groupby('Contract_Expiry').size().plot(kind='pie', label='')
```

```python
rating_counts = data['Rating'].value_counts().reset_index()
rating_counts.columns = ['Rating', 'Count']

# Create a scatter plot
plt.scatter(rating_counts['Rating'], rating_counts['Count'], alpha=0.5)
plt.title('Scatter Plot of Player Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()
```
